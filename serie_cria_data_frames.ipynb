Transformando uma Serie em DATAFRAME

    -Para transformar uma serie em um DataFrame utilizamos o concat() do Pandas;

        serie1 = Series(serie01)
        serie2 = Series(serie02)

        data_frame = concat([serie1, serie2], axis=1)

```python
import pandas
import matplotlib
from pandas import Series, concat
from random import randint
from string import ascii_uppercase
from pandas import Series

dict_exemplo = {k: randint(5, 50) for k in ascii_uppercase}

serie1 = Series(dict_exemplo)
serie2 = Series(dict_exemplo)

data_frame = concat([serie1, serie2], axis=1)
```

```python
data_frame
```

Podemos também transpor os dados do Data Frame:

    - data_frame.transpose()

```python
data_frame_transposto = data_frame.transpose()
data_frame_transposto
```
