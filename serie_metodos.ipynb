```python
from random import randint
from string import ascii_uppercase
import matplotlib

dict_exemplo = {k: randint(5, 50) for k in ascii_uppercase}
```

Estrutura da Serie:

```python
from pandas import Series

serie = Series(dict_exemplo)

#print serie
serie
```

Alguns métodos da Series:

- Buscando o valor minímo:

```python
serie.min()
```

- Buscando o valor máximo:


```python
serie.max()
```

- Média de valores:


```python
serie.mean()
```

- Mediana:

```python
serie.median()
```

Buscando a descrição dos dados apresentados na Serie:

serie. describe()

- count = número de valores;
- mean  = média de valores;
- std   = desvio padrão;
- max   = valor máximo;
- min   = valor mínimo;
- Quarts da serie (Porcentagens de valores da serie):
    25%
    50%
    75%
- dtype: tipo de dado dos valores da serie.

```python
serie.describe()
```

CRIANDO FILTROS :

Função descobrir dados apartir de um determinado valor:

- Filtrar valores maiores que valor  ->  serie[lambda x: x > valor_de_corte] 
* Essa função equivale à no python:
        
        def maior_que_17(x):
        
            return x> 17

```python
serie[lambda x: x > 17]
```

- Filtrar valores menores que valor  ->  serie[lambda x: x < valor_de_corte] 
* Essa função equivale à no python:
        
        def maior_que_17(x):
        
            return x <> 17

```python
serie[lambda x: x < 17]
```

ACESSAR VALORES PELO LABEL:

- Função loc[label] -> serie.loc[label]


```python
# Acessa o valor do label 'A' da serie

serie.loc['A']
```

ACESSAR VALORES PELA POSIÇÃO DO LABEL:

- Função iloc[posicao_do_label] -> serie.iloc[posicao_do_label]

```python
# Acessa o valor da primeira posição da serie atraves do primeiro label

serie.iloc[0]
```

Buscando uma faixa controlada de dados pela posição do label na serie:

-   serie.iloc[posição_inicial: posição_final];
-   serie.iloc[0: 10]

```python
serie.iloc[0: 10]
```

```python
serie_tratada.describe()
```
