#### Excluíndo Colunas de um DataFrame:

```python
import pandas as pd
```

```python
dados = pd.read_csv('aluguel.csv', sep = ';')
dados.head(10)
```

- Criando DataFrame Auxiliar para não afetar dataframe principal:

```python
dados_aux = pd.DataFrame(dados[['Valor Bruto', 'Valor Bruto M²', 'Tipo Agregado', 'Valor m2']])
dados_aux.head(10)
```

Método de exclusão de Colunas:

    - del df['coluna']:

```python
del dados_aux['Valor Bruto']

dados_aux.head(10)
```

Método de exclusão de Colunas:

    - df.pop('coluna'):

```python
dados_aux.pop('Valor Bruto M²')

dados_aux.head(10)
```

Método de exclusão de Colunas:

    - df.drop(['lista', 'de', 'colunas'], axis = 0|1, inplace = True|False)

    - O método 'drop()' para exclusão de dados serve tanto para excluir colunas, onde será passado a lista com os nomes das colunas (ou somente a coluna) que se deseja deletar,
    também é possível deletar a linha ao invés da coluna alterando o parametro de axis = 1 para axis = 0, é um método que consegue alterar diretamente o DataFrame através do paramêtro 'Inplace';

    #Obs: O método consegue excluir linhas apartir do index da linha, ou seja é necessário o index para remover a linha!

```python
dados.drop(['Valor Bruto', 'Valor Bruto M²'], axis=1, inplace=True)

dados.head(10)
```

- Salvando DataFrame:


```python
dados.to_csv('aluguel.csv', sep=';', index='False')
```
