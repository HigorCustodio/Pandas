Criando função que cria correlação do numero com o animal:

        - Criando co-relação onde é passado um numero para a função e esse numero devera retornar o animal correspondente ao numero:

```python
def numero_bicho(n, numeros=numeros):
    return numeros.loc[n]

print(numero_bicho(74))
```

        - Criando relação grupo - bicho:


```python
def grupo_bicho(bicho, grupos=grupos):
    return grupos.loc[bicho]

grupo_bicho(numero_bicho(74))
```

Por padrão o DataFrame desaparece com o 0 inicial do campo de valor, preciso do zero na frente do numero para gerar o numero de milhar;

        -Converter conjunto de numero para str;

        -Aplicar preenchimento para que sejam apresentados 4 números ao total no conjunto:


        numero_inicial = 123 (int)

        str(numero_inicial)

        >>>'123'

        <!-- str(conjunto_numeros).zfill(numero_total_que_deve_ter_casas) -->

        str(numero_inicial).zfill(4)

        >>>'0123'

Para separar a string com o valor em grupos:

        <!-- str(conjunto_numeros).zfill(numero_total_que_deve_ter_casas)[casas_1_grupo : valor_2_grupo]-->

        exemplos:

            str(numero_inicial).zfill(4)[:2]

            str(numero_inicial).zfill(4)[2:]

Buscando informações dentro do DataFrame   =   Query

        - df.query(label/filtro   ==   valor) <>  df.query('bicho_01 == "Touro"')

```python
df.query('bicho_01 == "Touro" ')
```

Tratando dados da busca:

1° passo: Posso pegar todos os resultados que tiverem Touro da coluna bicho_02                                  
        
         -> df.query('bicho_02 == "Touro"')

2° passo: Selecionar os valores que cairam na coluna bicho_01 para toda vez que deu "Touro" na coluna bicho_02  

         -> .bicho_01

3° passo: Agrupar o DataFrame pela coluna 'bicho_01' para mostrar só os resultados dela                       

         -> .groupby(df.bicho_01)

4° passo: Filtrar na coluna 'bicho_01' quantas vezes cada bicho apareceu para quando a coluna 'bicho_02 foi 'Touro'                                                                     

        -> .count()

código completo:

df.query('bicho_02 == "Touro"').bicho_01.groupby(df.bicho_01).count()

```python
df.query('bicho_02 == "Touro"').bicho_01.groupby(df.bicho_01).count()
```
