### Apresentando os valores nulos do DataFrame em forma de booleano :

-   Se utilizado df.isnunll(), será retornado um dataframe com valores booleanos onde o True serão as celulas que contém o Nan (ou vazio);
-   Se utilizado df.notnull(), será retornado um dataframe com valores booleanos onde o True serão as celulas que não contém valores nulos;

```python
dados_nulos = dados.isnull()
dados_nulos
```

```python
dados_notnull = dados.notnull()
dados_notnull
```

### Criando um novo DataFrame com somente os dados nulos e atribuindo ele ao dataframe existente:

- Podemos filtrar os valores nulos da coluna 'Valor' através do dados['Valor].isnull(), assim serão retornados uma Series contendo somente a coluna 'Valor' contendo o booleano da função isnull(): 
    - dados_att = dados['Valor'].isnull();

- Ou podemos sobrepor os dados no dataframe base usando algo semelhante a list comprehension do Python:
    - dados[dados['Valor'].isnull()];
    Desta forma o dataframe dados recebe o dataframe dados onde na coluna 'Valor' so existam  valores nulos;

```python
# dados[dados['Valor'].isnull()]
dados_att = dados['Valor'].isnull()
dados_att
```

### Retirando as linhas com valores nulos :

- Para retirar valores nulos do dataframe, podemos usar a função df.dropna(), essa função tem o paramêtro 'subset' onde esse paramêtro irá receber o nome da coluna a qual ele irá procurar os valores nulos 
em suas linhas:


        - dados.dropna(subset=['Valor'])


- Para utilizar o dataframe resultante desta função temos duas opções:
    Criar um novo dataframe e atribuir a ele os novos valores resultantes da função df.dropna():
        
        
        - dados_sem_nulos = dados.dropna(subset=['Valor'])


    Ou, Utilizar o paramêtro 'inplace' da propria função df.dropna(), onde as alterações serão feitas no proprio dataframe raiz:
        
        
        - dados.dropna(subset=['Valor'], inplace=True)

```python
dados_alterados = dados.shape[0]
dados.dropna(subset=['Valor'], inplace=True)
dados_nao_nulos = dados.shape[0]

dados_alterados - dados_nao_nulos
```

### Tratando dados nulos com condicionais:

```python
selecao = (dados['Tipo']=='Apartamento') & (dados['Condominio'].isnull())

dados = dados[~selecao]
dados
```

#### Podemos usar o fillna para substituir os valores nulos, onde se:
        
        - Usado somente o valor a ser colocado no lugar do NaN ele irá aplicar a mudança em todos os valores nulos do dataframe completo;

        - Para Colocar valores diferentes para os nulos em cada coluna podemos aplicar um dicionario contendo o nome da coluna como chave e o valor a ser colocado no lugar do nulo como valor;

```python
dados.fillna({'Condominio':0, 'IPTU':0}, inplace=True)
dados.info()
```
