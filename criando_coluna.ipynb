#### Criando novas colunas consumindo os dados das colunas existentes:

Dentro do Pandas é possível criar novas colunas tendo como base os dados inclusos nas colunas originais do DataFrame,

 é possível fazer operações matemáticas com os dados do mesmo tipo númerico ou até mesmo filtrar ou buscar valores strings;

```python
import pandas as pd
```

```python
dados = pd.read_csv('aluguel.csv', sep = ';')
dados.head(10)
```

```python
#Removendo coluna indesejada:
dados = dados.drop('Tipo Agregado', axis=1)
dados = dados.drop('Valor m2', axis=1)
```

Exemplos de operações matemáticas entre colunas:

Soma de valores:

```python
dados['Valor Bruto'] = dados['Valor'] + dados['Condominio'] + dados['IPTU']
dados.head(10)
```

Divisão de valores:

```python
dados['Valor m2'] = dados['Valor']/dados['Area']
dados.head(10)
```

* Ajustando casas decimais dos valores da coluna Valor M²

```python
dados['Valor m2'] = dados['Valor m2'].round(1)
dados.head(10)
```

Gerando o Valor Bruto do M²:

```python
dados['Valor Bruto M²'] = (dados['Valor Bruto'] / dados['Area']).round(1)
dados.head(10)
```

### Criando nova coluna através da manipulação de strings:

- Criando variável de controle:

```python
lista_casas = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
```

Método Apply:

        - O Método Apply do Pandas é um método criado para que seja possível aplicar determinada função a todas as linhas do DataFrame de forma índividual, é semelhante a um loop de repetição que irá passar uma vez por cada valor do DataFrame;

Função Lambda:

        -   É uma função do Python usada para criar uma variável de controle da função, onde é possível usar-la como se fosse o 'i' do 'for i in range', ou seja, ela irá representar cada celula do DataFrame, é util para se usar em funções do tipo ' lambda x: 'Casa' if x in lista_casas else 'Apartamento'';

```python
dados['Tipo Agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in lista_casas else 'Apartamento')
dados.head(10)
```

Salvando DataFrame:

```python
dados.to_csv('aluguel.csv', sep=';', index=False)
```

## Exemplo de criação de variável II

```python
alunos = pd.DataFrame({'Nome': ['Ary', 'Cátia', 'Denis', 'Beto', 'Bruna', 'Dara', 'Carlos', 'Alice'], 
                        'Sexo': ['M', 'F', 'M', 'M', 'F', 'F', 'M', 'F'], 
                        'Idade': [15, 27, 56, 32, 42, 21, 19, 35], 
                        'Notas': [7.5, 2.5, 5.0, 10, 8.2, 7, 6, 5.6]}, 
                        columns = ['Nome', 'Idade', 'Sexo', 'Notas'])

alunos.head(10)
```

- Cria uma variável que indica a diferença entre cada nota e a média das notas, indicando o quanto cada aluno ficou acima ou abaixo da média geral:
#### Usando - Lambda:

```python
alunos['Notas-Média(Notas)'] = alunos['Notas'] \
    .apply(lambda x: x - alunos['Notas'].mean())
alunos.head(10)
```

- Criando coluna onde as notas de cada aluno são comparadas com a nota média para saber o quanto ficou abaixo ou acima do valor médio:
### Métod manipulação de dados direta, Mais indicado por ter o tempo de processamento bem mais rápido que o Lambda:

```python

alunos['Notas-Média'] = \
    alunos.Notas - alunos.Notas.mean()

alunos.head(10)
```

- Usando if else(if else) com função Lambda:

```python
alunos['Faixa Etária'] = alunos['Idade'] \
    .apply(lambda x: 'Menor que 20 anos' if x < 20 
        else ('Entre 20 e 40 anos' if (x >= 20 and x <= 40) 
            else 'Maior que 40 anos'))

alunos.head(10)
```
